```python
df_cvd = pd.read_csv(os.path.join('/legodata/zhaoj/cvd_risk_time2/data/processed', 'exom_cvd.csv'))
df_fram = pd.read_csv(os.path.join('/legodata/zhaoj/cvd_risk_time2/genetic_data/for_framingham/processed/merged_fr_results.csv'))
```

```python
df_exome_cohort = df_exome.merge(df_fram[['predict', 'GRID']], how='inner', right_on='GRID', left_on='grid') 
print(len(df_exome_cohort))
df_exome_cohort.head()
```

```python
df_missing_frs = pd.Series(list(set(df_exome['grid']) - set(df_exome_cohort['grid'])))
print(len(df_missing_frs))
df_missing_frs.head()
```

```python
df_missing_frs.to_csv('/legodata/exome/missing_frs.csv', index=False)
```

```python
unique_case_lst = set(df_cvd['GRID'])
len(unique_case_lst)
```

```python
df_exome_cohort = df_exome
df_exome_cohort['CLASS'] = np.where(df_exome_cohort['grid'].isin(unique_case_lst), 1, 0)
df_exome_cohort.head()
```

```python
df_exome_cohort.to_csv('/legodata/exome/cvd_merged/TW_Liver.csv', index=False)
```
