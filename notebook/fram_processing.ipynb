```python
import pandas as pd
import os 

data_path = '/legodata/zhaoj/cvd_risk_time2/data_1_8/processed/'

df_merged_full_cohort = pd.read_csv('/legodata/zhaoj/cvd_risk_time2/data/processed/aggregate/df_merged_impute_final_0.csv')
df_fram_full_cohort = pd.read_csv('/legodata/zhaoj/cvd_risk_time2/data_1_8/for_framingham/processed/merged_framingham.csv')
```

```python
df_merged_full_cohort = df_merged_full_cohort.rename(columns={'Unnamed: 0': 'GRID'})
df_merged_full_cohort.head()
```

```python
df_fram_full_cohort.head()
```

```python
df_merged_full_cohort_fram = df_merged_full_cohort.merge(df_fram_full_cohort[['GRID', 'predict']], how='inner', on='GRID')
df_merged_full_cohort_fram['predict'] = df_merged_full_cohort_fram['predict'].astype('int64')
df_merged_full_cohort_fram.head()
```

```python
df_merged_full_cohort_fram.to_csv(os.path.join(data_path, 'df_full_cohort_0.csv'), index=False)
```
